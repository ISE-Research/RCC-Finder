```python
def time_interval(firstdate, seconddate):
    firstday = int(firstdate[8:10])
    firstmonth = int(firstdate[5:7])
    firstyear = int(firstdate[2:4])
    secondday = int(seconddate[8:10])
    secondmonth = int(seconddate[5:7])
    secondyear = int(seconddate[2:4]) 
    first = firstyear * 360 + (firstmonth - 1) * 30 + firstday
    second = secondyear * 360 + (secondmonth - 1) * 30 + secondday
    days = second - first
    return days
```

```python
#calculate whodo file_change score
import pandas as pd
import math
address = "results/flink/file_change_num.csv"
numresult_df = pd.read_csv(address)
address = "results/flink/file_change_num.csv"
timeresult_df = pd.read_csv(address)
address = "results/flink/file_change_num.csv"
expresult_df = pd.read_csv(address)
# for col in numresult_df.columns[1:]:
#     numresult_df[col].values[:] = 0
# for col in timeresult_df.columns[1:]:
#     timeresult_df[col].values[:] = 0
# for col in expresult_df.columns[1:]:
#     expresult_df[col].values[:] = 0
    
address = "D:/Master/semester3/experiments/features/flink/datedPullFile.csv"
file_df = pd.read_csv(address)
address = "D:/Master/semester3/experiments/features/flink/datedCommitFileAuthor.csv"
commit_df = pd.read_csv(address)
for i in range(numresult_df.shape[0]):
    number = numresult_df.loc[i, 'number']
    print(i)
    print(number)
    myfile_df = file_df[file_df.pull_number == number]
    if myfile_df.shape[0] == 0:
        continue
    date = myfile_df.iloc[0, 2]
    for j in range(myfile_df.shape[0]):
        mycommit_df = commit_df[commit_df.date < date]
        path = myfile_df.iloc[j, 1]
        newcommit_df = mycommit_df[mycommit_df.file_path==path]
        num_df = newcommit_df.groupby(['author_login'])['oid'].count()
        time_df = newcommit_df.groupby(['author_login'])['date'].max()
        for k in range(num_df.shape[0]):
            committer = num_df.index[k]
            num = num_df.iloc[k]
            max_value = time_df.iloc[k]
            numresult_df.loc[i, committer] += num
            lasttime = time_interval(max_value, date)
            if lasttime !=0:
                timeresult_df.loc[i, committer] += num / lasttime
            else:
                timeresult_df.loc[i, committer] += num
            expresult_df.loc[i, committer] += num / math.exp(lasttime/7)

numresult_df.to_csv("results/flink/file_change_num.csv", index = False)
timeresult_df.to_csv("results/flink/file_change_time.csv", index = False)
expresult_df.to_csv("results/flink/file_change_exp.csv", index = False)
timeresult_df
```

```python
#f_measure & precision & recall calculation
import pandas as pd
address = "results/spark/whodo_tops.csv"
tops_df = pd.read_csv(address)
address = "D:/Master/semester3/experiments/dataset/csv/github/apache/spark/2020-01-10/reviewer.csv"
reviewer_df = pd.read_csv(address, sep="|")
precision1 = 0
recall1 = 0
f_measure1 = 0
precision2 = 0
recall2 = 0
f_measure2 = 0
precision3 = 0
recall3 = 0
f_measure3 = 0
precision5 = 0
recall5 = 0
f_measure5 = 0
precision10 = 0
recall10 = 0
f_measure10 = 0


tops = tops_df.values.tolist()
for i in range(tops_df.shape[0]):
    number = tops_df.loc[i, 'number']
    myreviewer_df = reviewer_df[reviewer_df.pull_number == number]
    actual = myreviewer_df["reviewer_login"].tolist()
    recommended = tops[i]
    
    recommended1 = recommended[1:2]
    intersection_set = set.intersection(set(actual), set(recommended1))
    intersection_list = list(intersection_set)
    precision1 += len(intersection_list)/len(recommended1)
    p1 = len(intersection_list)/len(recommended1)
    recall1 += len(intersection_list)/len(actual)
    r1 = len(intersection_list)/len(actual)
    if (p1+r1) !=0:
        f_measure1 += 2*p1*r1/(p1+r1)
    
    recommended2 = recommended[1:3]
    intersection_set = set.intersection(set(actual), set(recommended2))
    intersection_list = list(intersection_set)
    precision2 += len(intersection_list)/len(recommended2)
    recall2 += len(intersection_list)/len(actual) 
    p2 = len(intersection_list)/len(recommended2)
    r2 = len(intersection_list)/len(actual) 
    if (p2+r2) !=0:
        f_measure2 += 2*p2*r2/(p2+r2)
    
    recommended3 = recommended[1:4]
    intersection_set = set.intersection(set(actual), set(recommended3))
    intersection_list = list(intersection_set)
    precision3 += len(intersection_list)/len(recommended3)
    recall3 += len(intersection_list)/len(actual) 
    p3 = len(intersection_list)/len(recommended3)
    r3 = len(intersection_list)/len(actual) 
    if (p3+r3) !=0:
        f_measure3 += 2*p3*r3/(p3+r3)
    
    recommended5 = recommended[1:6]
    intersection_set = set.intersection(set(actual), set(recommended5))
    intersection_list = list(intersection_set)
    precision5 += len(intersection_list)/len(recommended5)
    recall5 += len(intersection_list)/len(actual)
    p5 = len(intersection_list)/len(recommended5)
    r5 = len(intersection_list)/len(actual) 
    if (p5+r5) !=0:
        f_measure5 += 2*p5*r5/(p5+r5)
        
    recommended10 = recommended[1:11]
    intersection_set = set.intersection(set(actual), set(recommended10))
    intersection_list = list(intersection_set)
    precision10 += len(intersection_list)/len(recommended10)
    recall10 += len(intersection_list)/len(actual) 
    p10 = len(intersection_list)/len(recommended10)
    r10 = len(intersection_list)/len(actual) 
    if (p10+r10) !=0:
        f_measure10 += 2*p10*r10/(p10+r10)
    
print(precision1/tops_df.shape[0])
print(recall1/tops_df.shape[0])
print(f_measure1/tops_df.shape[0])
print(precision2/tops_df.shape[0])
print(recall2/tops_df.shape[0])
print(f_measure2/tops_df.shape[0])
print(precision3/tops_df.shape[0])
print(recall3/tops_df.shape[0])
print(f_measure3/tops_df.shape[0])
print(precision5/tops_df.shape[0])
print(recall5/tops_df.shape[0])
print(f_measure5/tops_df.shape[0])
print(precision10/tops_df.shape[0])
print(recall10/tops_df.shape[0])
print(f_measure10/tops_df.shape[0])
```
