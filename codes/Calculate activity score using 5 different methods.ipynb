```python
#activity score with different time coefficients
import pandas as pd
import math
address = "results/kafka/file_review_num.csv"
cuberesult_df = pd.read_csv(address)
address = "results/kafka/file_review_num.csv"
lnresult_df = pd.read_csv(address)
address = "results/kafka/file_review_num.csv"
forthresult_df = pd.read_csv(address)
address = "results/kafka/file_review_num.csv"
sqrtresult_df = pd.read_csv(address)
for col in cuberesult_df.columns[1:]:
    cuberesult_df[col].values[:] = 0
for col in lnresult_df.columns[1:]:
    lnresult_df[col].values[:] = 0
for col in forthresult_df.columns[1:]:
    forthresult_df[col].values[:] = 0
for col in sqrtresult_df.columns[1:]:
    sqrtresult_df[col].values[:] = 0
address = "dataset/csv/github/apache/kafka/2020-01-10/review.csv"
review_df = pd.read_csv(address)
address = "features/kafka/datedpullauthors.csv"
pull_df = pd.read_csv(address)
for i in range(cuberesult_df.shape[0]):
    number = cuberesult_df.loc[i, 'number']
    mypull_df = pull_df[pull_df.pull_number == number]
    date = mypull_df.iloc[0, 2]
    newdate = ""
    month = int(date[5:7])
    if month == 1:
        month = 12
        year = int(date[2:4])
        year -= 1
        newdate = date[:2] + str(year) + date[4:5] + str(month) + date[7:]
    else:
        month -= 1
        if month < 10:
            newdate = date[:5] + "0" + str(month) + date[7:]
        else:
            newdate = date[:5] + str(month) + date[7:]
    print(i)
    print(number)
    myreview_df = review_df[review_df.date < date]
    myreview_df = myreview_df[myreview_df.date > newdate]
    myreview_df = myreview_df.drop_duplicates(subset=['pull_number', 'reviewer_login'], keep="last")
    num_df = myreview_df.groupby(['reviewer_login'])['pull_number'].count()
    time_df = myreview_df.groupby(['reviewer_login'])['date'].max()
    for k in range(num_df.shape[0]):
        reviewer = num_df.index[k]
        num = num_df.iloc[k]
        max_value = time_df.iloc[k]
        lasttime = time_interval(max_value, date)
        numresult_df.loc[i, reviewer] = num
        lnresult_df.loc[i, reviewer] += num / math.log(lasttime+2)
        if lasttime !=0:
            sqrtresult_df.loc[i, reviewer] = num/math.sqrt(lasttime)
            cuberesult_df.loc[i, reviewer] = num/(lasttime**(1/3))
            forthresult_df.loc[i, reviewer] = num / (lasttime**(1/4))
        else: 
            sqrtresult_df.loc[i, reviewer] = num
            cuberesult_df.loc[i, reviewer] = num
            forthresult_df.loc[i, reviewer] = num
lnresult_df.to_csv("results/kafka/activity4_ln.csv", index = False)
cuberesult_df.to_csv("results/kafka/activity4_cube.csv", index = False)
forthresult_df.to_csv("results/kafka/activity4_4throot.csv", index = False)
sqrtresult_df.to_csv("results/kafka/activity4_sqrt.csv", index = False)
```
