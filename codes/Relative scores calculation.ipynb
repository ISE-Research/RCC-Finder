```python
#calculate relative file change score
import pandas as pd
import math
address = "results/flink/file_change_num.csv"
t2result_df = pd.read_csv(address)
address = "results/flink/file_change_num.csv"
sqrtresult_df = pd.read_csv(address)
address = "results/flink/file_change_num.csv"
relativeresult_df = pd.read_csv(address)
address = "results/flink/file_change_num.csv"
relativenumresult_df = pd.read_csv(address)
for col in t2result_df.columns[1:]:
    t2result_df[col].values[:] = 0
for col in sqrtresult_df.columns[1:]:
    sqrtresult_df[col].values[:] = 0
for col in relativeresult_df.columns[1:]:
    relativeresult_df[col].values[:] = 0
for col in relativenumresult_df.columns[1:]:
    relativenumresult_df[col].values[:] = 0
    
address = "features/flink/datedPullFile.csv"
file_df = pd.read_csv(address)
address = "features/flink/datedCommitFileAuthor.csv"
commit_df = pd.read_csv(address)
for i in range(relativenumresult_df.shape[0]):
    number = relativenumresult_df.loc[i, 'number']
    print(i)
    print(number)
    myfile_df = file_df[file_df.pull_number == number]
    if myfile_df.shape[0] == 0:
        continue
    date = myfile_df.iloc[0, 2]
    for j in range(myfile_df.shape[0]):
        mycommit_df = commit_df[commit_df.date < date]
        path = myfile_df.iloc[j, 1]
        newcommit_df = mycommit_df[mycommit_df.file_path==path]
        whole = newcommit_df.shape[0]
        num_df = newcommit_df.groupby(['author_login'])['oid'].count()
        time_df = newcommit_df.groupby(['author_login'])['date'].max()
        for k in range(num_df.shape[0]):
            committer = num_df.index[k]
            num = num_df.iloc[k]
            max_value = time_df.iloc[k]
            if whole != 0:
                relativenumresult_df.loc[i, committer] += num/whole
                lasttime = time_interval(max_value, date)
                if lasttime !=0:
                    relativeresult_df.loc[i, committer] += (num/whole) / lasttime
                    t2result_df.loc[i, committer] += (num/whole) / (lasttime^2)
                    sqrtresult_df.loc[i, committer] += (num/whole) / math.sqrt(lasttime)
                else:
                    relativeresult_df.loc[i, committer] += (num/whole) 
                    t2result_df.loc[i, committer] += (num/whole) 
                    sqrtresult_df.loc[i, committer] += (num/whole) 
                    
t2result_df.to_csv("results/flink/file_change_t2.csv", index = False)
sqrtresult_df.to_csv("results/flink/file_change_sqrt.csv", index = False)
relativeresult_df.to_csv("results/flink/file_change_relative.csv", index = False)
relativenumresult_df.to_csv("results/flink/file_change_relativenum.csv", index = False)
```
